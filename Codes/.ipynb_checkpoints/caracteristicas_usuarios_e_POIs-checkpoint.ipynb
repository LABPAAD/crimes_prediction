### Distribuições de frequência:

```python
dados = list(df.tweet_count.sort_values())

df_freq = pd.DataFrame(dados, columns=["Dados"])

#determina o intervalo de classes
intervalo_classe = 10000

# criar a tabela de frequência
tabela_freq = pd.DataFrame({"Intervalo de Classe": pd.cut(df_freq["Dados"], bins=range(min(df_freq["Dados"]), max(df_freq["Dados"]) + intervalo_classe, intervalo_classe), include_lowest=True)})
tabela_freq["Frequência"] = tabela_freq.groupby("Intervalo de Classe")["Intervalo de Classe"].transform("count")
tabela_freq["Frequência Relativa"] = tabela_freq["Frequência"] / len(df_freq)
tabela_freq["Porcentagem"] = tabela_freq["Frequência Relativa"] * 100
tabela_freq = tabela_freq.drop_duplicates().reset_index(drop=True)

# imprimir a tabela de frequência
print(tabela_freq)
```

### Tranformando dados quantitativos em valores anuais

```python
## tranforma a feature created_at para formato datetime
df['created_at'] = pd.to_datetime(df['created_at'])
```

```python
## adiciona uma nova feature no dataset com o ano de criação das contas
df['ano'] = df['created_at'].dt.year
```

```python
# Calculando o número de anos que o usuário está na plataforma
df['anos_ativo'] = round((pd.Timestamp('now') - df['created_at']).dt.days / 365)
```

```python
df["tweets_por_ano"] = df.tweet_count/df.anos_ativo
df["followers_por_ano"] = df.followers_count/df.anos_ativo
df["following_por_ano"] = df.following_count/df.anos_ativo

df.head()
```
