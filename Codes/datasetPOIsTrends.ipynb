```python
df = pd.read_csv('../Datasets/trends/trends2022.csv')

# Converta a coluna 'date' para o tipo datetime
df['date'] = pd.to_datetime(df['date'])

# Defina a coluna 'date' como índice
df.set_index('date', inplace=True)

# Agrupe por mês e calcule as proporções mensais
df_anual = df.resample('Y').sum()/52  # Agrupa por ano e soma os valores e divide pela quantidade

df_anual = df_anual.transpose()
```
