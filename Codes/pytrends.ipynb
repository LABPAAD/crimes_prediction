```python
import pandas as pd
import numpy as np
from pytrends.request import TrendReq
from glob import glob
import threading
```

```python
lista_dps = glob("../Datasets/SSP/2022/*")
lista_dps.sort()
```

```python
#lista com termos de pesquisa

termos = []
for dp in lista_dps:
    item = dp.split("-")
    item = item[2]
    item = item.join(" ")+item+"São Paulo"
    item = item.lstrip()
    item = item.lower()
    termos.append(item)
termos
```

### Usando Threads

```python
# Crie um objeto TrendReq
pytrends = TrendReq()

# Defina os termos de pesquisa
termos_de_pesquisa = termos

# Função que obtém as tendências para um termo de pesquisa
def obter_tendencias(termo, index):
    pytrends = TrendReq(hl='pt-BR', tz=180)
    pytrends.build_payload([termo], timeframe='2022-01-01 2022-12-31', geo='BR')
    dados_de_tendencias[index] = (termo, pytrends.interest_over_time())

# Dicionário para armazenar os dados das tendências para cada termo de pesquisa
dados_de_tendencias = {}

# Lista para armazenar as threads
threads = []

# Criar e iniciar as threads
for index, termo in enumerate(termos_de_pesquisa):
    thread = threading.Thread(target=obter_tendencias, args=(termo, index))
    threads.append(thread)
    thread.start()

# Aguardar até que todas as threads sejam concluídas
for thread in threads:
    thread.join()

# Criar os DataFrames na ordem da lista original de termos de pesquisa
dataframes = []
for index in range(len(termos_de_pesquisa)):
    termo, dados = dados_de_tendencias[index]
    df = pd.DataFrame(dados)
    dataframes.append(df)

# Concatenar os DataFrames em um único DataFrame
df_final = pd.concat(dataframes, axis=1)
```

```python
df_final = df_final.drop(columns="isPartial")
```
